## Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
metadata = pd.read_csv("data/metadata.tsv", sep='\t')
metadata.head(5)
```

```python
col = ["gene_id", "gene_name", "gene_biotype", ] + metadata['File accession'].tolist()
TPM = None

for i in range(metadata.shape[0]):
    tmp = pd.read_csv(f"data/{metadata.loc[i, 'File accession']}.tsv", 
                     sep='\t')
    tmp = tmp[tmp.gene_id.str.contains('ENSG', na=False, regex=True)]
    tmp.reset_index(drop=True, inplace=True)
    tmp = pd.DataFrame(tmp[["gene_id", "TPM"]])
    tmp.index = tmp.gene_id
    tmp.drop(["gene_id"], axis=1, inplace=True)
    tmp.columns = [metadata.loc[i, 'File accession']]
    if TPM is None:
        TPM = tmp.copy()
    else:
        TPM = pd.concat([TPM, tmp], axis=1)
TPM
```

```python
gene_info = pd.read_csv("gencode_v29_gene_metadata.txt", sep='\t')
gene_info.index = gene_info.gid
gene_info.drop(["gid"], axis=1 , inplace=True)
TPM = pd.concat([TPM, gene_info], axis=1)
cols = TPM.columns[-3:].tolist() + TPM.columns[:-3].tolist()
TPM = TPM[cols]
TPM.to_csv("TPM_matrix.csv")
TPM
```

```python
TPM = TPM[np.logical_or(TPM.biotype == "protein_coding", 
                        TPM.biotype == "pseudogene",
                        TPM.biotype == "lncRNA")]
TPM.to_csv("TPM_ployA_matrix.csv")
TPM
```

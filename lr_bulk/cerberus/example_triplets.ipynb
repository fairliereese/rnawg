```python
import cerberus
import pandas as pd
```

```python
tss = [1,1,4,2,1]
ic = [4,1,2,2,1]
tes = [1,2,1,2,1]
df = pd.DataFrame()
df['n_tss'] = tss
df['n_ic'] = ic
df['n_tes'] = tes
```

```python
df = cerberus.compute_splicing_ratio(df)
```

```python
df
```

```python
df['total'] = df.n_tss+df.n_tes+df.splicing_ratio
df['tss_ratio'] = df.n_tss/df.total
df['top_ratio'] = df.splicing_ratio/df.total
df['tes_ratio'] = df.n_tes/df.total
```

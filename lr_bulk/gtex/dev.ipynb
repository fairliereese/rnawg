## Weird stuff going on w/ gtex stuff? 
* Yunzhe says that some transcripts from the transcript novelty file (which is output directly from the gtex gtf) don't match to the cerberus ids
* This is because all of the ones that are removed b/w the two files are from novel genes

```python
def get_gtex_gene_nov(df):
    df['gene_novelty'] = 'Known'
    df.loc[df.gene_id.str.contains('chr'), 'gene_novelty'] = 'Novel'
    return df

def get_gtex_transcript_nov(df):
    df['transcript_novelty'] = 'Novel'
    df.loc[df.transcript_id.str.contains('ENST'), 'transcript_novelty'] = 'Known'
    return df
```

```python
f1 = 'gtex_to_cerberus_id.tsv'
f2 = 'gtex_transcript_novelty.tsv'

df1 = pd.read_csv(f1, sep='\t')
df2 = pd.read_csv(f2, sep='\t')
```

```python
df1.head()
```

```python
df2.transcript_id.head().values
```

```python
# get list of transcripts that don't have a cerberus mapping
tids = list(set(df2.transcript_id.tolist())-set(df1.original_transcript_id.tolist()))
len(tids)
```

```python
gtf = 'flair_filter_transcripts.gtf'
df = pr.read_gtf(gtf, as_df=True)
df = get_gtex_gene_nov(df)
df = get_gtex_transcript_nov(df)
```

```python
temp = df.loc[df.transcript_id.isin(tids)]
temp = temp[['transcript_id', 'gene_novelty']].drop_duplicates()
temp.groupby('gene_novelty').count()
```

```python
import pyranges as pr
```

```python
df = pr.read_gtf('flair_filter_transcripts.gtf')
```

```python
df = df.as_df()
```

```python
df.head()
```

```python
genes = df[['gene_id']].drop_duplicates()
genes['gene_novelty'] = 'Novel'
genes.loc[genes.gene_id.str.contains('ENSG'), 'gene_novelty'] = 'Known'
genes.groupby('gene_novelty').count()
```

```python
ts = df[['transcript_id']].drop_duplicates()
ts['transcript_novelty'] = 'Novel'
ts.loc[ts.transcript_id.str.contains('ENST'), 'transcript_novelty'] = 'Known'
ts.groupby('transcript_novelty').count()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import seaborn as sns
import sys
import os

p = os.path.dirname(os.path.dirname(os.getcwd()))
sys.path.append(p)

from scripts.utils import *
from scripts.plotting import *
```

```python
df = pd.read_csv('../../lr_bulk/talon/human_talon_abundance.tsv', sep='\t')
df = get_tpm_table(df, sample='cell_line', how='gene', save=True)
```

```python
top_genes = pd.DataFrame()
for c in df.columns:
    df = df.sort_values(by=c, ascending=False)
#     print(df.head())
    temp = pd.DataFrame(df.iloc[:10, :].copy(deep=True).index.to_list(), 
                        columns=[c])
    top_genes = pd.concat([top_genes, temp], axis=1)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import seaborn as sns
import sys
import os

p = os.path.dirname(os.path.dirname(os.getcwd()))
sys.path.append(p)

from scripts.utils import *
from scripts.plotting import *
```

```python
df = pd.read_csv('../../lr_bulk/talon/human_talon_abundance.tsv', sep='\t')
tpm_df = get_tpm_table(df, sample='cell_line', how='gene', save=True)
```

```python
temp = df[['annot_gene_id', 'annot_gene_name']].drop_duplicates()
tpm_df = tpm_df.merge(temp, how='left', 
                      left_index=True, right_on='annot_gene_id')
tpm_df.set_index('annot_gene_name', inplace=True)
tpm_df.drop('annot_gene_id', axis=1, inplace=True)
```

```python
top_genes = pd.DataFrame()
for c in tpm_df.columns:
    tpm_df = tpm_df.sort_values(by=c, ascending=False)
#     print(df.head())
    temp = pd.DataFrame(tpm_df.iloc[:10, :].copy(deep=True).index.to_list(), 
                        columns=[c])
    top_genes = pd.concat([top_genes, temp], axis=1)
```
